# TAVI 연구 로드맵
**Thermal Accessibility Vulnerability Index**

> 업데이트: 2026-04-10 | 버전: TAVI_v1

## 연구 방향

### Plan A (우선 실행)
공간환경 변수(NDVI, 불투수면, 건물높이 등)가 역세권 Thermal Catchment 감소율을 설명한다

```
종속변수: reduction_pct (역별 × 시간대별)
독립변수: NDVI, 불투수면, 건물높이, 하천거리, 가로수
분석: OLS / GWR
```

### Plan B (확장)
H×E×V 프레임워크 → TAVI = Hazard × Exposure × Vulnerability

```
H: catchment zone 평균 UTCI
E: reduction_pct
V: 고령인구 비율 + 자가용 미보유
```
