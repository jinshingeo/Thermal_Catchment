## 연구 핵심 주장

> **"기존 Catchment는 물리적 이동 제약만 반영한다.
> 우리는 열환경(H×E)과 인구사회학적 취약성(V)을 통합한
> TAVI 프레임워크를 제안하고,
> 이것이 폭염 시 실제 대중교통 이용 패턴과 대응됨을 보인다."**

### 프레임워크 구조 (IPCC AR6 Risk = H×E×V)

```
TAVI(역) = reduction_pct × vulnerability_ratio

  H (Hazard)      : SOLWEIG 기반 UTCI 열환경
  E (Exposure)    : Thermal Catchment reduction_pct
                    = (Classic - Thermal) / Classic × 100
  V (Vulnerability): (65세+ 인구 + 14세- 인구) / 전체 인구
                    (역 Catchment 내 행정동 가중 평균)
```

### Kar et al.(2023) 연결

| Kar et al. | TAVI |
|-----------|------|
| Hard constraint (물리적 장벽) | H × E (UTCI 기반 열환경 장벽) |
| Soft constraint (개인 인식) | V (취약 집단일수록 열 패널티 강화) |
